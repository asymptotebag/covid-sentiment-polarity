```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_root = '/data/'
```

```python
def sentiment_type(x):
    if x == 0:
        return 'Negative'
    elif x == 1:
        return 'Neutral'
    elif x == 2:
        return 'Positive'
    
def tweet_len(x):
    return len(x.split())
```

```python
df_final = pd.read_csv(os.path.join(data_root ,'multiclass_dataset_final.csv'))
df_final = df_final.drop(['Unnamed: 0.1', 'Unnamed: 0', 'id'], axis=1)
df_final['Sentiment'] = df_final['multiclass_sentiment'].apply(sentiment_type)
df_final['Tweet Length'] = df_final['text'].apply(tweet_len)
df_final.head(2)
```

```python
duplicate = df_final[df_final['text'].duplicated()]
len(duplicate)
```

```python
sns.countplot(x='Sentiment', data=df_final, alpha = 0.5)
plt.title('Sentiment Classes Size')
plt.show()
```

```python
sns.displot(data=df_final, x="norm_sentiment", hue="Sentiment", multiple="stack", bins=33, alpha = 0.5)
plt.title('Distribution of Normalized Sentiment Scores')
plt.show()
```

```python
sns.catplot(data=df_final, x="Sentiment", y="norm_sentiment", kind="boxen")
plt.title('Values of Normalized Sentiment Scores by Sentiment Type')
```

```python
sns.displot(data=df_final, x="Tweet Length", hue="Sentiment", bins=50, alpha = 0.4)
plt.title('Distribution of Tweet Length by Sentiment')
plt.show()
```

```python
sns.catplot(data=df_final, x="Sentiment", y="Tweet Length", kind="boxen")
plt.title('Distribution of Tweet Length by Sentiment')
```
